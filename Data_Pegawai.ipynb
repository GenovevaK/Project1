```python
import csv
from tabulate import tabulate
import pandas as pd
from google.colab import files


pegawai_db = {
    "001": ("Genoveva", "Manager", "Keuangan", "50000000", "01-01-2020", "Tetap", "genoveva@bca.co.id", "08123456789"),
    "002": ("Alex", "Developer", "IT", "40000000", "15-03-2021", "Tetap", "alex@bca.co.id", "08123456788"),
    "003": ("Fika", "Data Scientist", "Analisis Data", "40000000", "20-05-2023", "Kontrak", "fika@bca.co.id", "08123456787"),
    "004": ("Vanessa", "Teller", "Operasional", "30000000", "10-07-2014", "Tetap", "vanessa@bca.co.id", "08123456786"),
    "005": ("Ayes", "Developer", "IT", "40000000", "01-08-2024", "Tetap", "ayes@bca.co.id", "08123456785"),
    "006": ("Rindang", "Digital Banking Officer", "IT", "40000000", "01-09-2023", "Tetap", "rindang@bca.co.id", "08123458985")
}


admin_password = "GZ123"


def format_gaji(gaji):
    return "{:,.0f}".format(int(gaji)).replace(',', '.')


def tambah_data_pegawai():
    id_pegawai = input("Masukkan ID Pegawai: ")
    if id_pegawai in pegawai_db:
        print(f"Pegawai dengan ID {id_pegawai} sudah ada.")
        return


    nama = input("Masukkan Nama Pegawai: ")
    jabatan = input("Masukkan Jabatan Pegawai: ")
    departemen = input("Masukkan Departemen Pegawai: ")
    gaji = input("Masukkan Gaji Pegawai: ")
    tanggal_bergabung = input("Masukkan Tanggal Bergabung (DD-MM-YYYY): ")
    status = input("Masukkan Status Pegawai: ")
    email = input("Masukkan Email Pegawai: ")
    telepon = input("Masukkan Nomor Telepon Pegawai: ")


    pegawai_db[id_pegawai] = (nama, jabatan, departemen, gaji, tanggal_bergabung, status, email, telepon)
    print(f"Data pegawai {nama} berhasil ditambahkan.")


def tampilkan_data_pegawai(id_pegawai):
    if not pegawai_db:
        print("Tidak ada data pegawai.")
        return

    table = []
    for key, value in pegawai_db.items():
        if key == id_pegawai:
            formatted_gaji = format_gaji(value[3])
        else:
            formatted_gaji = "*****"
        table.append([key, *value[:3], formatted_gaji, *value[4:]])

    print(tabulate(table, headers=["ID", "Nama", "Jabatan", "Departemen", "Gaji", "Tanggal Bergabung", "Status", "Email", "Telepon"], tablefmt="fancy_grid"))


def tampilkan_semua_data_pegawai():
    if not pegawai_db:
        print("Tidak ada data pegawai.")
        return

    table = []
    for key, value in pegawai_db.items():
        formatted_gaji = format_gaji(value[3])
        table.append([key, *value[:3], formatted_gaji, *value[4:]])

    print(tabulate(table, headers=["ID", "Nama", "Jabatan", "Departemen", "Gaji", "Tanggal Bergabung", "Status", "Email", "Telepon"], tablefmt="fancy_grid"))


def edit_data_pegawai():
    id_pegawai = input("Masukkan ID Pegawai yang ingin diedit: ")
    if id_pegawai in pegawai_db:
        print("Pilih kolom yang ingin diedit:")
        print("1. Nama")
        print("2. Jabatan")
        print("3. Departemen")
        print("4. Gaji")
        print("5. Tanggal Bergabung")
        print("6. Status")
        print("7. Email")
        print("8. Telepon")
        pilihan = input("Masukkan nomor kolom yang ingin diedit (1-8): ")

        if pilihan == "1":
            nama = input(f"Nama ({pegawai_db[id_pegawai][0]}): ") or pegawai_db[id_pegawai][0]
            pegawai_db[id_pegawai] = (nama, *pegawai_db[id_pegawai][1:])
        elif pilihan == "2":
            jabatan = input(f"Jabatan ({pegawai_db[id_pegawai][1]}): ") or pegawai_db[id_pegawai][1]
            pegawai_db[id_pegawai] = (pegawai_db[id_pegawai][0], jabatan, *pegawai_db[id_pegawai][2:])
        elif pilihan == "3":
            departemen = input(f"Departemen ({pegawai_db[id_pegawai][2]}): ") or pegawai_db[id_pegawai][2]
            pegawai_db[id_pegawai] = (*pegawai_db[id_pegawai][:2], departemen, *pegawai_db[id_pegawai][3:])
        elif pilihan == "4":
            gaji = input(f"Gaji ({pegawai_db[id_pegawai][3]}): ") or pegawai_db[id_pegawai][3]
            pegawai_db[id_pegawai] = (*pegawai_db[id_pegawai][:3], gaji, *pegawai_db[id_pegawai][4:])
        elif pilihan == "5":
            tanggal_bergabung = input(f"Tanggal Bergabung ({pegawai_db[id_pegawai][4]}): ") or pegawai_db[id_pegawai][4]
            pegawai_db[id_pegawai] = (*pegawai_db[id_pegawai][:4], tanggal_bergabung, *pegawai_db[id_pegawai][5:])
        elif pilihan == "6":
            status = input(f"Status ({pegawai_db[id_pegawai][5]}): ") or pegawai_db[id_pegawai][5]
            pegawai_db[id_pegawai] = (*pegawai_db[id_pegawai][:5], status, *pegawai_db[id_pegawai][6:])
        elif pilihan == "7":
            email = input(f"Email ({pegawai_db[id_pegawai][6]}): ") or pegawai_db[id_pegawai][6]
            pegawai_db[id_pegawai] = (*pegawai_db[id_pegawai][:6], email, pegawai_db[id_pegawai][7])
        elif pilihan == "8":
            telepon = input(f"Telepon ({pegawai_db[id_pegawai][7]}): ") or pegawai_db[id_pegawai][7]
            pegawai_db[id_pegawai] = (*pegawai_db[id_pegawai][:7], telepon)
        else:
            print("Pilihan tidak valid!")

        print(f"Data pegawai {id_pegawai} berhasil diperbarui.")
    else:
        print(f"Pegawai dengan ID {id_pegawai} tidak ditemukan.")


def hapus_data_pegawai():
    id_pegawai = input("Masukkan ID Pegawai yang ingin dihapus: ")
    if id_pegawai in pegawai_db:
        del pegawai_db[id_pegawai]
        print(f"Data pegawai dengan ID {id_pegawai} telah dihapus.")
    else:
        print(f"Pegawai dengan ID {id_pegawai} tidak ditemukan.")


def export_data_pegawai():
    try:

        data_pegawai = {
            "ID": [],
            "Nama": [],
            "Jabatan": [],
            "Departemen": [],
            "Gaji": [],
            "Tanggal Bergabung": [],
            "Status": [],
            "Email": [],
            "Telepon": []
        }

        for key, value in pegawai_db.items():
            data_pegawai["ID"].append(key)
            data_pegawai["Nama"].append(value[0])
            data_pegawai["Jabatan"].append(value[1])
            data_pegawai["Departemen"].append(value[2])
            data_pegawai["Gaji"].append(value[3])
            data_pegawai["Tanggal Bergabung"].append(value[4])
            data_pegawai["Status"].append(value[5])
            data_pegawai["Email"].append(value[6])
            data_pegawai["Telepon"].append(str(value[7]))


        df = pd.DataFrame(data_pegawai)


        df.to_csv('data_pegawai.csv', index=False)
        print("Data pegawai berhasil diekspor ke data_pegawai.csv")


        files.download('data_pegawai.csv')
    except Exception as e:
        print(f"Terjadi kesalahan saat mengekspor data: {e}")


def menu_admin():
    while True:
        print("\n=== Menu Admin ===")
        print("1. Tambah Data Pegawai")
        print("2. Edit Data Pegawai")
        print("3. Hapus Data Pegawai")
        print("4. Lihat Data Pegawai")
        print("5. Export Data Pegawai")
        print("6. Logout")
        pilihan = input("Pilih opsi (1-6): ")

        if pilihan == "1":
            tambah_data_pegawai()
        elif pilihan == "2":
            edit_data_pegawai()
        elif pilihan == "3":
            hapus_data_pegawai()
        elif pilihan == "4":
            tampilkan_semua_data_pegawai()
        elif pilihan == "5":
            export_data_pegawai()
        elif pilihan == "6":
            print("Logout berhasil!")
            break
        else:
            print("Pilihan tidak valid!")


def login():
    while True:
        print("\n=== Login ===")
        role = input("Masukkan role (admin/user): ").lower()
        if role == "admin":
            password = input("Masukkan password admin: ")
            if password == admin_password:
                print("Login sebagai Admin.")
                menu_admin()
                break
            else:
                print("Password salah, coba lagi.")
        elif role == "user":
            id_pegawai = input("Masukkan ID Pegawai: ")
            if id_pegawai in pegawai_db:
                print("Login sebagai User.")
                menu_user(id_pegawai)
                break
            else:
                print("ID Pegawai tidak valid, coba lagi.")
        else:
            print("Role tidak valid, coba lagi.")


def menu_user(id_pegawai):
    while True:
        print("\n=== Menu User ===")
        print("1. Lihat Profil Saya")
        print("2. Lihat Daftar Pegawai")
        print("3. Logout")
        pilihan = input("Pilih opsi (1-3): ")

        if pilihan == "1":
            tampilkan_profil_pegawai(id_pegawai)
        elif pilihan == "2":
            tampilkan_data_pegawai(id_pegawai)
        elif pilihan == "3":
            print("Logout berhasil!")
            break
        else:
            print("Pilihan tidak valid!")


def tampilkan_profil_pegawai(id_pegawai):
    if id_pegawai in pegawai_db:
        data = pegawai_db[id_pegawai]
        print(f"ID: {id_pegawai}, Nama: {data[0]}, Jabatan: {data[1]}, Gaji: {format_gaji(data[3])}, Tanggal Bergabung: {data[4]}, Status: {data[5]}, Email: {data[6]}, Telepon: {data[7]}")
    else:
        print(f"Pegawai dengan ID {id_pegawai} tidak ditemukan.")


if __name__ == "__main__":
    login()
```
